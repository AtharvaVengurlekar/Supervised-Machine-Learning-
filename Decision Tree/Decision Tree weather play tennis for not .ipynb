```python
import pandas as pd
```

```python
data = pd.read_csv(r'C:\Users\Atharva Vengurlekar\Desktop\Machine Learning\Decision Tree\tennis.csv')
data
```

```python
#Convert into the numeric form 
from sklearn.preprocessing import LabelEncoder
```

```python
outlook = LabelEncoder()
temp = LabelEncoder()
humidity = LabelEncoder()
windy = LabelEncoder()
play = LabelEncoder()
```

```python
data['outlook']=outlook.fit_transform(data['outlook'])
data['temp']=outlook.fit_transform(data['temp'])
data['humidity']=outlook.fit_transform(data['humidity'])
data['windy']=outlook.fit_transform(data['windy'])
data['play']=outlook.fit_transform(data['play'])
data
```

```python
#Seprate the independent and dependent variables
feature_col = ['outlook','temp','humidity','windy']
x = data[feature_col]
y = data.play
```

```python
x
```

```python
y
```

```python
#split to train test data 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifer = DecisionTreeClassifier(criterion = 'gini')
```

```python
classifer.fit(x_train,y_train)
```

```python
classifer.predict(x_test)
```

```python
classifer.score(x_test,y_test)
```

```python
from sklearn import tree
tree.plot_tree(classifer)
```
